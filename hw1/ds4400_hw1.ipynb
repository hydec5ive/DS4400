```python
import numpy as np

# Set seed for reproducibility (optional)
np.random.seed(42)

# Generate 3 random 3x3 matrices with integers from -10 to 10
matrices = []
for i in range(3):
    M = np.random.randint(-10, 11, size=(3, 3))
    matrices.append(M)
    print(f"Matrix {i+1}:")
    print("Original Matrix")
    print(M)
    print

    det = np.linalg.det(M)
    print(f"Determinant: {det:.4f}")
    print()
    
    if abs(det) > 0.0001:
        inv_M = np.linalg.inv(M)
        print("Inverse of Matrix:")
        print(inv_M)
        print()
        
        product = M @ inv_M
        print("Verification (M × inverse_Matrix):")
        print(np.round(product, 10))
        print()
    else:
        print("This matrix is SINGULAR - No inverse exists!")
        print()
```

```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('kc_house_data.csv')

# Remove id, date, zipcode
df = df.drop(columns=['id', 'date', 'zipcode'])

# Separate features and response
features = [col for col in df.columns if col != 'price']

# Part 1: Compute statistics
stats = df[features].agg(['mean', 'min', 'max', 'var']).T
stats.columns = ['Average', 'Min', 'Max', 'Variance']

# Part 2: Correlation with price
corr = df[features].corrwith(df['price']).sort_values(ascending=False)

# Display results
print("PART 1: Statistics")
print(stats)
print("\nPART 2: Correlation with Price")
print(corr)
```
