```python
import numpy as np
import pandas as pd

np.random.seed(42)

# Generate 3 random 3x3 matrices from -10 to 10
matrices = []
for i in range(3):
    M = np.random.randint(-10, 11, size=(3, 3))
    matrices.append(M)
    print(f"Matrix {i+1}:")
    print("Original Matrix")
    print(M)
    print()

# Generate determinant for each matrix
    det = np.linalg.det(M)
    print(f"Determinant: {det:.4f}")
    print()
    
    if abs(det) > 0.0001:
        inv_M = np.linalg.inv(M)
        print("Inverse of Matrix:")
        print(inv_M)
        print()
        
        product = M @ inv_M
        print("Verification (M × inverse_Matrix):")
        print(np.round(product, 10))
        print()
    else:
        print("Does not have inverse")
```

```python
# Question 5: Average, variance, and correlation
# Load data
df = pd.read_csv('kc_house_data.csv')

# Remove id, date, zipcode
df = df.drop(columns=['id', 'date', 'zipcode'])

# Select feature but price
features = [col for col in df.columns if col != 'price']

# Part 1: Compute statistics
stats = df[features].agg(['mean', 'min', 'max', 'var']).T.round(3)

# Results
print("PART 1: Statistics")
print(stats)
print(f"\nHighest average feature:{stats['mean'].idxmax()} ({stats['mean'].max()})")
print(f"Lowest average feature:{stats['mean'].idxmin()} ({stats['mean'].min()})")
print(f"Highest variance feature:{stats['var'].idxmax()} ({stats['var'].max()})")
print(f"Lowest variance feature:{stats['var'].idxmin()}({stats['var'].min()})")
```

```python
# Part 2: Correlation with price
corr = df[features].corrwith(df['price']).round(3).sort_values(ascending=False)

# Results
print(f"\nPART 2: Correlation with Price\n")
print(corr)

# Filter only positive correlations
positive_corr = corr[corr > 0]

# List them
print(f"\nFeatures: {list(positive_corr.index)}")

# Highest positive correlation
print(f"Highest positive correlation: {corr.idxmax()} ({corr.max()})")
```

```python
# Part 3: Negative correlations
negative_corr = corr[corr < 0]
print(f"\nPART 3: Negative Correlations")
print(f"Number of negative correlations: {len(negative_corr)}")
```
