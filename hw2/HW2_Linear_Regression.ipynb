Assume that we have $N=30$ samples of a feature $X$ representing the age of a person, and response $Y$ representing the income of the person. We are given the following:

- Sample mean of $X$ is $\bar{X}=50$; sample standard deviation of $X$ is $\sigma_X=10$;
- Sample mean of $Y$ is $\bar{Y} =100$; sample standard deviation of $Y$ is $\sigma_y=20$;
- The correlation coefficient between $X$ and $Y$ is $\rho =0.8$.

1. Compute the $\theta_0$ and $\theta_1$ coefficients of the least square linear regression model $h_{\theta} (x) =\theta_0 + \theta_1 x$ that predicts response $Y$ based on feature $X$
2. Change the correlation coefficient to $\rho=-0.8$. How do the parameters of the least square linear regression model change?

```python
N = 30
X_bar = 50
sigma_X = 10
Y_bar = 100
sigma_Y = 20
p = 0.8

# Part 1: Calculate theta_1 and theta_0 with p = 0.8
theta_1 = p * (sigma_Y / sigma_X)
theta_0 = Y_bar - theta_1 * X_bar

print("Part 1: ρ = 0.8")
print(f"θ₁ (slope) = ρ × (σ_Y / σ_X) = {p} × ({sigma_Y} / {sigma_X}) = {theta_1:.4f}")
print(f"θ₀ (intercept) = Ȳ - θ₁ × X̄ = {Y_bar} - {theta_1:.4f} × {X_bar} = {theta_0:.4f}")
print(f"\nLinear regression model: Y = {theta_0:.4f} + {theta_1:.4f}X\n")

# Part 2: Calculate with p = -0.8
p_neg = -0.8
theta_1_neg = p_neg * (sigma_Y / sigma_X)
theta_0_neg = Y_bar - theta_1_neg * X_bar

print("Part 2: ρ = -0.8")
print(f"θ₁ (slope) = ρ × (σ_Y / σ_X) = {p_neg} × ({sigma_Y} / {sigma_X}) = {theta_1_neg:.4f}")
print(f"θ₀ (intercept) = Ȳ - θ₁ × X̄ = {Y_bar} - {theta_1_neg:.4f} × {X_bar} = {theta_0_neg:.4f}")
print(f"\nLinear regression model: Y = {theta_0_neg:.4f} + {theta_1_neg:.4f}X\n")

print("Interpretation:")
print(f"When ρ changes from {p} to {p_neg}:")
print(f"θ₁ changes from {theta_1:.4f} to {theta_1_neg:.4f} (sign flips, magnitude stays same)")
print(f"θ₀ changes from {theta_0:.4f} to {theta_0_neg:.4f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Set random seed for reproducibility
np.random.seed(42)
```

## Data Loading and Preprocessing

Load the data and apply the required transformations:
1. Remove 'id', 'date', and 'zipcode' columns
2. Scale features to mean 0 and standard deviation 1
3. Divide price by 1000

```python
# Load the datasets
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

print("Training data shape:", train_df.shape)
print("Testing data shape:", test_df.shape)
print("\nColumns:", train_df.columns.tolist())
```

```python
# Data Cleaning/Transformation:
drop_cols = ['Unnamed: 0', 'id', 'date', 'zipcode']
train = train_df.drop([c for c in drop_cols if c in train_df.columns], axis = 1)
test = test_df.drop([c for c in drop_cols if c in test_df.columns], axis = 1)
```

```python
# Scale the price by dividing by 1000
train['price'] = train['price'] / 1000
test['price'] = test['price'] / 1000
```

```python
# Prepare the data for modeling
X_train = train.drop(columns=['price'])
y_train = train['price']
X_test = test.drop(columns=['price'])
y_test = test['price']
```

```python
# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train the linear regression model
model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
# Coefficients
feature_names = X_train.columns.tolist()
print(f"Coefficient for each feature:\n")
coe_feature = []
for name, coef in zip(feature_names, model.coef_):
    coe_feature.append((name, coef))

pd.DataFrame(coe_feature, columns=['Feature', 'Coefficient'])
```

```python
# Evaluate the model on the training dataset
y_train_pred = model.predict(X_train_scaled)
train_mse = mean_squared_error(y_train, y_train_pred)
train_r2 = r2_score(y_train, y_train_pred)
print(f"Training MSE: {train_mse:.4f}")
print(f"Training R²: {train_r2:.4f}")
```

# 2.2 Evaluate the model on the testing set. Report the MSE and $R^2$ metrics on the testing set.

```python
# Evaluate the model on the testing dataset
y_test_pred = model.predict(X_test_scaled)
test_mse = mean_squared_error(y_test, y_test_pred)
test_r2 = r2_score(y_test, y_test_pred)
print(f"Testing MSE: {test_mse:.4f}")
print(f"Testing R²: {test_r2:.4f}")
```

# [C] Problem 3:  Implementing closed-form solution for linear regression (15 points) 

In this problem, you will implement your own linear regression model, using the closed-form solution we derived in class. You will also
compare your model with the one trained with the package in Problem 2 on the same house price prediction dataset.

- Implement the closed-from solution for multiple linear regression using matrix operations and train a model on the training set. Write
a function to predict the response on a new testing point.
- Compare the models given by your implementation with those trained in Problem 2 by the Python packages. Report the MSE and $R^2$
metrics for the models you implemented on both training and testing sets and compare these metrics to the ones given by the package
implementation from Problem 2. Discuss if the results of your implementation are similar to those of the package.

```python
# θ = (XᵀX)⁻¹Xᵀy 
def fit_closed_form(X, y):
    # Add column of 1s for intercept: (N, d) -> (N, d+1)
    X_b = np.hstack([np.ones((X.shape[0], 1)), X])
    # Solve XᵀX θ = Xᵀy 
    return np.linalg.solve(X_b.T @ X_b, X_b.T @ y)

def predict_closed_form(X, theta):
    # Add 1s column to match the theta vector
    X_b = np.hstack([np.ones((X.shape[0], 1)), X])
    # prediction = θ₀ + θ₁x₁ + θ₂x₂ + ... for each house
    return X_b @ theta

theta = fit_closed_form(X_train_scaled, y_train)

y_train_pred = predict_closed_form(X_train_scaled, theta)
y_test_pred = predict_closed_form(X_test_scaled, theta)

print(f"Train MSE: {round(mean_squared_error(y_train, y_train_pred), 4)}")
print(f"Train R2:  {round(r2_score(y_train, y_train_pred), 4)}")
print(f"Test MSE:  {round(mean_squared_error(y_test, y_test_pred), 4)}")
print(f"Test R2:   {round(r2_score(y_test, y_test_pred), 4)}")
```

The results are virtually identical. Both methods give Train MSE ≈ 31416, Train R² ≈ 0.727, Test MSE ≈ 58835, and Test R² ≈ 0.647.

# [C] Problem 5:  Gradient descent (20 points)

In this problem, you will implement your own gradient descent algorithm and apply it to linear regression on the same house prediction dataset.

1. Write code for gradient descent for training linear regression using the algorithm from class.
2. Vary the value of the learning rate (at least 3 different values $\alpha \in \{0.01,0.1,0.5\}$) and report the value of the model parameter $\theta$ after different number of iterations (10, 50, and 100). Include in a table the MSE and $R^2$ metrics on the training and testing set for the different number of iterations and different learning rates. You can choose more values of the learning rates to observe how the
behavior of the algorithm changes.
3. Write some observations about the behavior of the algorithm: How do the metrics change with different learning rates; How many
iterations are needed; Does the algorithm converge to the optimal solution, etc.

```python
def gradient_descent(X, y, alpha, num_iterations):
    # Add column of 1s for intercept: (N, d) -> (N, d+1)
    X_b = np.hstack([np.ones((X.shape[0], 1)), X])
    N = X_b.shape[0]
    # Start with all weights at zero
    theta = np.zeros(X_b.shape[1])
    
    for i in range(num_iterations):
        y_pred = X_b @ theta                       
        error = y_pred - y                         
        gradient = (2 / N) * (X_b.T @ error)       
        theta = theta - alpha * gradient           
    
    return theta

# Results table: 3 learning rates × 3 iteration counts
results = []

for alpha in [0.01, 0.1, 0.5]:
    for n_iter in [10, 50, 100]:
        theta = gradient_descent(X_train_scaled, y_train.values, alpha, n_iter)
        
        y_train_pred = predict_closed_form(X_train_scaled, theta)
        y_test_pred = predict_closed_form(X_test_scaled, theta)
        
        results.append({
            'Alpha': alpha,
            'Iters': n_iter,
            'Train MSE': round(mean_squared_error(y_train, y_train_pred), 2),
            'Train R2': round(r2_score(y_train, y_train_pred), 4),
            'Test MSE': round(mean_squared_error(y_test, y_test_pred), 2),
            'Test R2': round(r2_score(y_test, y_test_pred), 4)})

pd.DataFrame(results)
```

```python
print("Theta values for each alpha and iteration count:\n")

for alpha in [0.01, 0.1, 0.5]:
    print(f"Alpha = {alpha}:")
    for n_iter in [10, 50, 100]:
        t = gradient_descent(X_train_scaled, y_train.values, alpha, n_iter)
        print(f"Iter {n_iter:>3}: {list(np.round(t, 4))}")
    print()

theta_opt = fit_closed_form(X_train_scaled, y_train.values)
print(f"Optimal (closed-form): {list(np.round(theta_opt, 4))}")
```

- α = 0.01 — too slow. After 10 iterations the model has negative R², meaning it's worse than just predicting the mean price for every house. By 100 iterations it's at R² ≈ 0.68, which is better but still below the optimal 0.727.

- α = 0.1 — just right. By 50 iterations we're at R² ≈ 0.727 and MSE ≈ 31427, nearly matching the closed-form. By 100 iterations it's essentially identical (MSE ≈ 31416). 

- α = 0.5 — too big. The MSE explodes to numbers with dozens of digits. Each step overshoots the minimum, and the next step overshoots even more, and it spirals out of control exponentially.

# [A/C] Problem 6: Ridge regularization (20 points)

In this problem, you will derive the optimal parameters for ridge regression and train ridge regression models with different regularization levels. In ridge regression, the loss function includes a regularization term:

$J(\theta) = \sum_{i=1}^N(h_{\theta}(x_i)-y_i)^2 + \lambda \sum_{j=1}^d \theta_j^2$

1. **[A]** Write the derivation of the closed form solution for parameter $\theta$ that minimizes the loss function $J(\theta)$ in ridge regression.
2. **[C]** Modify your implementation from Problem 5 to implement ridge regression with gradient descent.
3. **[C]** Simulate $N=1000$ values of random variable $X_i$, distributed uniformly on interval $[-2,2]$. Simulate the values of random variable       $Y_i = 1 + 2X_i + e_i$, where $e_i$ is drawn from a Gaussian distribution $N(0, 2)$. Fit this data with linear regression, and also with ridge regression
for different values of $\lambda \in \{1,10,100,1000,10000\}$. Print the slope, the MSE values, and the $R^2$ statistic for each case and write down some observations. What happens as the regularization parameter $\lambda$ increases?

```python
# Full derivation in PDF
print("Closed-form solution for ridge: θ = X^T y / X^T X + λI")
```

```python
def ridge_gradient_descent(X, y, lam, alpha, num_iterations):
    X_b = np.hstack([np.ones((X.shape[0], 1)), X])
    N = X_b.shape[0]
    theta = np.zeros(X_b.shape[1])
    
    for i in range(num_iterations):
        y_pred = X_b @ theta
        error = y_pred - y
        gradient = (2 / N) * (X_b.T @ error)
        gradient[1:] += (2 * lam / N) * theta[1:]
        theta = theta - alpha * gradient
    
    return theta
```

```python
N = 1000
X_sim = np.random.uniform(-2, 2, N).reshape(-1, 1)
noise = np.random.normal(0, np.sqrt(2), N)
y_sim = 1 + 2 * X_sim.flatten() + noise
```

```python
results = []
for lam in [0, 1, 10, 100, 1000, 10000]:
    theta = ridge_gradient_descent(X_sim, y_sim, lam, alpha=0.01, num_iterations=1000)
    
    y_pred = predict_closed_form(X_sim, theta)
    mse = mean_squared_error(y_sim, y_pred)
    r2 = r2_score(y_sim, y_pred)
    
    results.append({
        'Lambda': lam,
        'Intercept': theta[0],
        'Slope': theta[1],
        'MSE': mse,
        'R2': r2})

pd.DataFrame(results)
```

λ = 0 (no regularization): Intercept ≈ 1, slope ≈ 2, matching the true parameters. This is ordinary least squares.
As λ increases: The slope shrinks toward 0 while the intercept stays near 1. MSE increases and R² decreases because the model is forced to use a smaller slope.
At very large λ (10000): The slope is nearly 0. The model predicts roughly the same value for every house, ignoring X entirely. R² drops close to 0.
